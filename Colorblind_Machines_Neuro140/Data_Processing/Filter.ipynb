```python
# Libraries for Filtering
!pip install colorblind
from colorblind import colorblind
import cv2
import glob, os.path
from os import listdir,makedirs
from os.path import isfile,join

# Source Folder
path = '/content/drive/MyDrive/Harvard_BAI/Colorblind_Machines_Neuro140/data/CIFAR-10-images-master/train_rgb' 

# Destination Folder Deuteranopia
dstpath_deu = '/content/drive/MyDrive/Harvard_BAI/Colorblind_Machines_Neuro140/data/CIFAR-10-images-master/train_deu' 
# Destination Folder Grayscale
dstpath_gy = '/content/drive/MyDrive/Harvard_BAI/Colorblind_Machines_Neuro140/data/CIFAR-10-images-master/train_gy'
# Destination Folder Protanopia
dstpath_pro = '/content/drive/MyDrive/Harvard_BAI/Colorblind_Machines_Neuro140/data/CIFAR-10-images-master/train_pro' 

# Ensure all dst directories exist/created if don't exist
dstpath_list = [dstpath_deu,dstpath_gy, dstpath_pro]

# RGB Photos to filter for other experiemtns
files=listdir(path)

```

### Need to create seperate loops to filter images Cannot create a single loop -- runtime too long

```python
# Create Deuteranopia
for dstpath in dstpath_list:
  if dstpath  == dstpath_deu:
    for image in files:
      if not (os.path.exists(dstpath +'/%s'%image)):
        try:
            img = cv2.imread(os.path.join(path,image))
            deu = colorblind.simulate_colorblindness(img, colorblind_type='deuteranopia')
            dstPath = join(dstpath,image)
            cv2.imwrite(dstPath,deu)
        except:
            print ("{} is not converted".format(image))
      for fil in glob.glob("*.jpg"):
        try:
          image = cv2.imread(fil) 
          deu_image = colorblind.simulate_colorblindness(os.path.join(path,image), colorblind_type='deuteranopia') 
          cv2.imwrite(os.path.join(dstpath,fil),deu_image)
        except:
          print('{} is not converted').format(fil)
  
  
  
```

```python
# Create Grayscale
for dstpath in dstpath_list:
  if dstpath == dstpath_gy:
    for image in files:
      if not (os.path.exists(dstpath +'/%s'%image)):
        try:
            img = cv2.imread(os.path.join(path,image))
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            dstPath = join(dstpath,image)
            cv2.imwrite(dstPath,gray)
        except:
            print ("{} is not converted".format(image))
      for fil in glob.glob("*.jpg"):
        try:
          image = cv2.imread(fil) 
          gray_image = cv2.cvtColor(os.path.join(path,image), cv2.COLOR_BGR2GRAY) 
          cv2.imwrite(os.path.join(dstpath,fil),gray_image)
        except:
          print('{} is not converted').format(fil)
```

```python
# Create Protanopia
for dstpath in dstpath_list:
  if dstpath == dstpath_pro:
      for image in files:
        if not (os.path.exists(dstpath +'/%s'%image)):
          try:
              img = cv2.imread(os.path.join(path,image))
              pro = colorblind.simulate_colorblindness(img, colorblind_type='protanopia')
              dstPath = join(dstpath,image)
              cv2.imwrite(dstPath,pro)
          except:
              print ("{} is not converted".format(image))
        for fil in glob.glob("*.jpg"):
          try:
            image = cv2.imread(fil) 
            pro_image = colorblind.simulate_colorblindness(os.path.join(path,image), colorblind_type='protanopia') 
            cv2.imwrite(os.path.join(dstpath,fil),pro_image)
          except:
            print('{} is not converted').format(fil)
```
